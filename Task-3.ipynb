```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv('C:\\Users\\MAGANTI SAI\\Documents\\Main Flow\\householdtask3.csv')
```

```python
display(data.head(10))
```

```python
#Scatter plot along with  year aganist own
plt.scatter(data['year'],data['own'])

#adding title to the plot
plt.title("Scatter Plot")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#line chart
plt.plot(data['year'])
plt.plot(data['own'])

#adding title to the plot
plt.title("Line Plot")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#Bar Chart -representing in rectangular boxe's
plt.bar(data['year'],data['own'])

#adding title to the plot
plt.title("Bar Chart")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

#title
plt.title("Histogram")

#adding legends
plt.show()
```
